```python
# Ans 2

A scatter plot is a visual representation of how two variables relate to each other.
```

```python
# Ans 2
import matplotlib.pyplot as plt
import numpy as np 
```

```python
np.random.seed(3)
```

```python
x = 3 + np.random.normal(0, 2, 50)
```

```python
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x , y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Random Graph')
plt.show()
```

```python
# Ans 3
The subplot() function takes three arguments that describes the layout of the figure. 
```

```python
#For line 1
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2,3,1)
plt.plot(x,y)

#  for line 2

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2,3,2)
plt.plot(x,y)

# For line 3
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2,3,3)
plt.plot(x,y)

# for line 4

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,3,4)
plt.plot(x,y)

plt.show()
```

```python
# Ans 4

The bar() function takes arguments that describes the layout of the bars.
```

```python
# solution of 4

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.show()
```

```python
# solution of 4(horizontal bar graph)

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company,profit)
plt.show()
```

```python
# Ans 5
A box plot which is also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum.

# use 
box plot are a great chart to use when showing the distribution of data point across a selected measure.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
figure = plt.figure(figsize =(10, 8))
plt.boxplot(box1)
plt.show()
```

```python
figure = plt.figure(figsize =(10, 8))
plt.boxplot(box2)
plt.show()
```
